# Matplotlib ScatterMatrix and Seaborn PairGrid

If you worked with real world data, you know how a lot of it is just junk. 

It's not just junk in the sense of data that needs cleansing.  A lot of data is just noise. It has no bearing on the inferences you want to make. 

I spend a lot of time cleaning then evaluating data before it's fed into model.  I use ScatterMatrix/PairGrid extensively for this.  

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
```

```python
df = pd.read_csv('data.csv')
```

```python
df.shape
```

```python
df.columns
```

## Matplotlib ScatterMatrix


```python
matplotlib.rc('axes',edgecolor='w')

disp_df = df[['count1', 'count2', 'total','regular', 'overage1','overage2']]

scatter_cells = pd.plotting.scatter_matrix(disp_df, 
                                           figsize=[36.0, 36.0])
```

## Seaborn PairGrid 

```python
g = sns.PairGrid(disp_df)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

## Seaborn PairGrid with KDE

This is the one I like to use most.

Instead of the stock PairGrid, the diagonal histogram plot and lower plots are now Kernel Density Estimation plots.  Good for visualizing the 'contour' of data grouping.

```python
g = sns.PairGrid(disp_df, diag_sharey=False)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
```
