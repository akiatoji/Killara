```python
import tensorflow as tf
import tensorflow_datasets as tfds

%matplotlib inline
import matplotlib.pyplot as plt   
```

```python
ds, ds_info = tfds.load('eurosat', split='train', with_info=True)
ds = ds.skip(2).take(1)
```

```python
i = 1
for d in ds:  
  src_img= d["image"]

  img_array = tf.image.rot90(src_img)
  img_array = img_array.numpy().astype('uint8')
  plt.subplot(211)
  plt.imshow(img_array, cmap='gray')
  plt.subplot(212)
  plt.imshow(src_img, cmap='gray')
  
plt.show()
```
